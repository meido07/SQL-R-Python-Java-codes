```python
# Adding some visualizations for group 1 (those who have tried study schedule have)
#https://seaborn.pydata.org/examples/index.html    # like GGplot in R
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
# Adding disttribution plot for group 1 (those who have tried study schedule have)
vis1 = sns.distplot(Stat1["referral_score"], bins = 30 )
vis1
```

```python
# Adding disttribution plot for group 2 (those who have not tried study schedule have)
vis2 = sns.distplot(Stat2["referral_score"], bins = 30 )
vis2
```

```python
# Group 1 (those who have tried study schedule have)
vis3 = sns.lmplot(data=F1, x='referral_score', y= 'user_id', 
                  fit_reg=False, hue='referral_score', size = 10, 
                  scatter_kws={"s":200})
```
